### Exploratory Data Analysis I - Second Approach 

In this section we are introducing the data and answer the following questions in the section below this statement: Why are emissions higher in the daytime? Why are concentrations higher at night and lower in the daytime?
> The command 'pd.read_csv'. allows us create a new table by reading the file located in the filename 

```python
data = pd.read_csv('exploratorium_bay_start_2022_2023.csv')
data.head(10)
```

Let's ignore this code below. It cleans the data frame a little more, and makes it more understandable.

```python
data.dropna(inplace = True)
new = data['local_timestamp'].str.split('-', n = 3, expand = True)
data['Year'] = new[0]
data['Month'] = new[1]
data['day'] = new[2]
data_2 = data[['Year', 'Month', 'day', 'node_id', 'co_ppm', 'co_ppm_QC_level', 'co2_ppm', 'co2_ppm_QC_level']]
new2 = data_2['day'].str.split(' ', n = 2, expand = True)
data['Day'] = new2[0]
data['pacific_time'] = new2[1]
emissions_df = data.loc[:, :]
```

```python
emissions_df.head(10)
```

As we can see our data frame has a lot variables, and it could be somehow complicated to read and use it to answer our questions. 

Let's select the columns we will need to answer our questions. We can use '[]' to select the column we want, and if we want to select multiples columns we can use '[[colum1, colum2]]' 

```python
emissions_df_2 = emissions_df[['Year', 'Month', 'Day', 'pacific_time','node_id', 'co_ppm', 'co_ppm_QC_level', 'co2_ppm', 'co2_ppm_QC_level']]
emissions_df_2
```

We can also notice that in the botton of our data frame we have a total of numeber of rows and columns. Thanks to this we can know how many rows and columns are we dealing with. Now what if we only want to display the first 10 rows. So, Let's use the function 'head()'

```python
emissions_df_2.head(10)
```

Now what if we want to sort our data frame by a specific column. We use the function 'sort_values' that takes 2 parameters. First, the name of the column and two 'ascending' that we can set it to True or False depending on how we want our values to be sorted.

```python
emissions_df.sort_values('co2_ppm', ascending = False).head(10)
```

Now we can use a groupby operation. This '.groupby()' operation generates mini sub-DataFrames. Each subframe contains all rows that correspond to a particular year. 

Let's use groupby in our 'pacific_time' columnn. In addition, let's use the aggregation method '.agg()', so we can summarize each DataFrameGroupBy object into one aggregated row per subframe. 

```python
(emissions_df.groupby('Year')
 .agg(max)
 .sort_values('co_ppm', ascending = True))
```

```python
(emissions_df.groupby('pacific_time')
 .agg(max)
 .sort_values('co2_ppm', ascending = False))
```

```python
(emissions_df.groupby('co2_ppm')
 .agg(max)
 .sort_values('co2_ppm', ascending = False).tail(10))
```

```python
(emissions_df.groupby('co2_ppm')
 .agg(max)
 .sort_values('co2_ppm', ascending = False).head(10))
```
