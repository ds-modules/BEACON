```python
#Loading Dataset
gd_data = Table.read_table('data/Dan Mini Elementary School_Glass Fire.csv')
gd_data = gd_data.drop("datetime","epoch", "julian_day","co2_ppm_QC_level", "o3_ppm", "o3_ppm_QC_level", "PM_2.5_ug/m3", "PM_2.5_ug/m3_QC_level")
gd_data= gd_data.with_column("lat", 38.1498075923)
gd_data = gd_data.with_column("lng",-122.258882174)

def assign_value(x):
    if x < 440:
        return 0.3
    elif x < 500:
        return 0.6
    elif x < 525:
        return 0.8
    elif x < 550:
        return 1.5
    else:
        return 0.1
    
gd_data['weights'] = gd_data.apply(assign_value, 'co2_ppm')
gd_data

#Convert to pandas and get the date and time
df_1 = gd_data.to_df()
df_1[['date', 'time']] = df_1['local_timestamp'].str.split(' ', expand=True)
df_1.drop('local_timestamp', axis=1, inplace=True)
df_1

#Now work on creating a pushing the latitude and longitude thing
lat = df_1['lat'].values 
lon = df_1['lng'].values 
w = df_1['weights'].values

call_locs = np.vstack((lat, lon,w)).transpose().tolist() 
call_locs[:5]


# Create a Heatmap with the call data.
heatmap = HeatMap(call_locs, radius = 10) 

#Creating a map for dan elementary
dan_coords = (38.1498075923, -122.258882174) 
dan_map = folium.Map(location= dan_coords, zoom_start=7) 

# Add it to your Berkeley map.
dan_map.add_child(heatmap)


```

```python
## Getting the node tbl - code from Glass Fires.py
def co2(start,end,*nodes):
    full = pd.DataFrame()
    for node in nodes:
        print(node)
        parameter = 'co2_corrected_avg_t_drift_applied-level-2'
        filename="http://128.32.208.8/node/"+node+"/measurements_all/csv?name=Supersite&interval=60&variables="+parameter+"&start="+start+"&end="+end
        data  = pd.read_csv(filename, on_bad_lines="skip",index_col=[0], parse_dates=[0]) 
        data = data.rename(columns={parameter:node})
        data = data.drop(['epoch','node_id','node_file_id','datetime'],axis = 1)
        data.replace(-999, np.nan, inplace=True)
        data.replace(-111, np.nan, inplace=True)
        full = pd.merge(full,data, how = 'outer',left_index=True,right_index=True) 
    return full
```

```python
#Pulling the data - you can change the start and end date, as well as the node IDs
co2_bay1 = co2('2020-09-15%2000:00:00','2020-10-27%2000:00:00','4','13','14','15','17','20','21','22','25','26','28')
co2_bay2 = co2('2020-09-15%2000:00:00','2020-10-27%2000:00:00','31','33','37','38','41','42','43','45','48','54')
co2_bay3 = co2('2020-09-15%2000:00:00','2020-10-27%2000:00:00','58','59','60','62','63')
```
